```python
frame = pd.DataFrame(np.arange(16).reshape(4,-1), 
                         columns = ['c1', 'c2', 'c3', 'c4'],
                         index = ['r1', 'r2', 'r3', 'r4'])
print(frame['c1'])
print(frame['c1']['r1']) 
print(frame[['c1']])
#print(frame[['c1']]['r1'])  #-> error!!
print(frame[['c1']]['c1'])
print(frame[['c1', 'c3']])
#print(frame[['c1', 'c3']]['r1'])  #-> error!!
print(frame[['c1', 'c3']]['c1'])

#
print('---------------------------')
print(frame['r1':'r2'])
print(frame['c1':'c2'])
#loc
print(frame.loc['r1', 'c1'])
print(frame.loc[['r1'], 'c1'])
print(frame.loc[:, 'c1':'c2'])
print(frame.loc['r1', :])
#iloc 
print(frame.iloc[0, 0])
print(frame.iloc[[0], 0])
print(frame.iloc[:, :1])
```

### get_dummies
명목형 변수의 경우 one-hot encoding을 해주는 경우가 있음.  

- one-hot(원핫)인코딩이란? 단 하나의 값만 True이고 나머지는 모두 False인 인코딩을 말한다.
즉, 1개만 Hot(True)이고 나머지는 Cold(False)이다.
예를들면 [0, 0, 0, 0, 1]이다. 5번째(Zero-based 인덱스이므로 4)만 1이고 나머지는 0이다.
- 여러 개로 분류하는 분류 문제에서
타켓 클래스 값을 정수로 사용할 수도 있고 One-Hot 인코딩한 값을 사용할 수도 있다.
연산 편의성, 다양한 모델 적용, 정확도 향상 등의 이유로 인코딩/디코딩한다.
- https://wikidocs.net/22647

  지역
0 서울
1 부산
2 대전

서울 1 0 0
부산 0 1 0
대전 0 0 1


서울 부산 대전
  1    0    0
  0    1    0
  0    0    1

```python
df = pd.DataFrame({'col1': [10, 20, 30],
                   'col2': ['a', 'b', 'a']})
df
```
